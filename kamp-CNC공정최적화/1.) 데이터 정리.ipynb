```python
import os 
import glob 
import pandas as pd
import numpy as np
from numpy import array

pd.set_option('display.max_columns', None)

import warnings 
warnings.filterwarnings(action='ignore')
```

```python
path = 'C:/Users/user/Desktop/연습작업물/kamp-CNC공정최적화\dataset/CNC 비식별화 원본데이터_1209/CNC Virtual Data set _v2'
df_list = glob.glob(path +'\*.csv') # 공정 실험 데이터 업로드 1 ~ 25번
df_list.remove(df_list[-1])  

#dfs = []

#for file in df_list:
#    df_load = pd.read_csv(file)
#    dfs.append(df_load)
    
#df = pd.concat(dfs).reset_index(drop=True)

#### 공정 실험 데이터 통합
dfs = []

for i in range(0,len(df_list)):
    df_experiment = pd.read_csv(df_list[i])
    df_experiment['index'] = i+1 # experiment 데이터 정보 추가
    dfs.append(df_experiment)
    
df = pd.concat(dfs).reset_index(drop=True)
df = df.drop_duplicates()

#### 공정 정보 
df_info = pd.read_csv('train.csv')  

df.shape, df_info.shape
```

#### 실험 데이터 정보 분류

```python
df_info['tool_condition'] = df_info['tool_condition'].replace({'unworn':0,'worn':1})

#### 1.) 공정  & 육안 식별 완료
Pass = df_info[(df_info['machining_finalized'] == 'yes') & (df_info['passed_visual_inspection'] == 'yes')]

#### 2.) 공정 완료 & 육안 식별 미완료
Fault = df_info[(df_info['machining_finalized'] == 'yes') & (df_info['passed_visual_inspection'] == 'no')]

#### 3.) 공정 미완료
Incomplete = df_info[df_info['machining_finalized'] == 'no']

print('Normal Counts : ',len(Pass))
print('Fault Counts : ',len(Fault))
print('Incomplete Counts : ',len(Incomplete))
print('\n')

#### 실험 데이터 집단 라벨링
Pass['Labeling'] = 0
Fault['Labeling'] = 1
Incomplete['Labeling'] = 2

#### 데이터 재정렬
df_info_reorganized = pd.concat([Pass,Fault,Incomplete],axis=0).sort_values(by='No')

Pass_Experiments = Pass['No'].tolist()
df_normal = df[df['index'].isin(Pass_Experiments) == True]
df_normal['Score'] = 0

Fault_Experiments = Fault['No'].tolist()
df_fault = df[df['index'].isin(Fault_Experiments) == True]
df_fault['Score'] = 1

Incomplete_Experiments = Incomplete['No'].tolist()
df_incomplete = df[df['index'].isin(Incomplete_Experiments) == True]
df_incomplete['Score'] = 2

print('Machining 완료 & Visual Inspection 통과 O --> Normal',len(df_normal))
print('Machining 완료 & Visual Inspection 통과 X --> Fault',len(df_fault))
print('Machining & Visual Inspection 작업 X --> Inomplete',len(df_incomplete))
print('\n')

data = pd.concat([df_normal,df_fault],axis=0).reset_index(drop=True)
data_test = df_incomplete.reset_index(drop=True)

print('최종 학습&검증 데이터] / [예측 데이터] 완성')
data.shape,data_test.shape
```

```python
data.to_csv('Train_Valid.csv',index=False)
```
